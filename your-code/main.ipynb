## Challenge 1
There is a box with 10 white balls, 12 red balls and 8 black balls. Calculate the probability of:
* Taking a white ball out.
* Taking a white ball on first extraction and a black ball out on second extraction without replacing the first one.
* Taking a white ball on first extraction and a black ball out on second extraction replacing the first one.
* Taking 3 balls and them being all different colours.

```python
white_balls = 10
red_balls = 12 
black_balls = 8

total_balls = white_balls + red_balls + black_balls

# Taking a white ball out
print(f"Probability of taking a white ball out {white_balls / total_balls}")
print()

# Taking a white ball on first extraction and a black ball out on second extraction without replacing the first one.
first_white = white_balls / total_balls
second_black = black_balls / (total_balls - 1)
print("Probability of Taking a white ball on first extraction and a black ball out on second extraction without replacing the first one.")
print(first_white * second_black)
print()


# Taking a white ball on first extraction and a black ball out on second extraction replacing the first one.
first_white = white_balls / total_balls
second_black = black_balls / total_balls
print("Probability of Taking a white ball on first extraction and a black ball out on second extraction replacing the first one.")
print(first_white * second_black)
print()


# Taking 3 balls and them being all different colours.
# w + r + b = 10/30 * 12/29 * 8/28
# w + b + r = 10/30 * 8/29 * 12/28
# b + w + r = 8/30 * 10/29 * 12/28
# b + r + w 
# r + w + b
# r + b + w
print("Probability of Taking 3 balls and them being all different colours.")
different_colors = (10/30 * 12/29 * 8/28) * 6
print(different_colors)
```

## Challenge 2

You want to go on picnic on Saturday. You hate rain, and would know to plan your weekend. To help you make a decision, you gather the following data about rainy days:

* 40% of all days are cloudy.
* About 50% of cloudy days are rainy. 
* About 20% of non-cloudy days are rainy.

What is the chance of rain during next Saturday?

```python
P_cloudy = 0.4
P_not_cloudy = 1 - 0.4
P_rain_given_cloudy = 0.5
P_rain_given_not_cloudy = 0.2

P_rain = P_rain_given_not_cloudy * P_not_cloudy + P_rain_given_cloudy * P_cloudy
P_rain
```
